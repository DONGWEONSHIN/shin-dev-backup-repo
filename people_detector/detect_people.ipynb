````python
import torch
import cv2
import numpy as np
````

````python
# YOLOv5 ëª¨ë¸ ë¡œë“œ
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
````

````python
# ì´ë¯¸ì§€ ë¶ˆëŸ¬ì˜¤ê¸°
image_path = 'classroom.jpg'  # ë¶„ì„í•  ì‚¬ì§„ ê²½ë¡œ
img = cv2.imread(image_path)
````

````python
# YOLO ëª¨ë¸ ì‹¤í–‰
results = model(img)
````

````python
# ê°ì§€ëœ ê°ì²´ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
detections = results.pred[0]
````

````python
# 'person' í´ë˜ìŠ¤ë§Œ í•„í„°ë§ (YOLOv5ì—ì„œ ì‚¬ëŒì˜ í´ë˜ìŠ¤ IDëŠ” 0)
people = [det for det in detections if int(det[-1]) == 0]
````

````python
# ê°ì§€ëœ ì‚¬ëŒ ìˆ˜ ì¶œë ¥
print(f"Detected People: {len(people)}")
````

````python
# ê°ì§€ëœ ì¸ì› ìˆ˜
person_count = len(people)
````

````python
# ê°ì§€ëœ ê°ì²´ ì‹œê°í™”
for det in people:
    x1, y1, x2, y2, conf, cls = det.cpu().numpy()  # ê°ì§€ëœ ê°ì²´ ì •ë³´
    cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
````

````python
# ì´ë¯¸ì§€ í¬ê¸° í™•ì¸
height, width, _ = img.shape

# í•˜ë‹¨ ì¤‘ì•™ ì¢Œí‘œ ê³„ì‚°
text_position = (width // 2 - 100, height - 50)

# ì¸ì› ìˆ˜ë¥¼ ì´ë¯¸ì§€ í•˜ë‹¨ ì¤‘ì•™ì— ì¶œë ¥
cv2.putText(img, f"Total Count: {person_count}", text_position, 
            cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 255), 3, cv2.LINE_AA)
````

````python
# ê²°ê³¼ ì´ë¯¸ì§€ ì €ì¥ ë° í‘œì‹œ
cv2.imwrite('output.jpg', img)
cv2.imshow("Detected People", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
````
